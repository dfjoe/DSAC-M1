## 몬테카를로 방법(Monte- Carlo method)
무작위 추출된 난수(Random Nomber)를 이용하여 함수의 값을 계산하는 통계학적 방법  
수치적분이나 최적화를 통해 널리 사용

### 몬테카를로 분석
random 숫자 발생을 이용하여 원주율 pi를 시뮬레이션을 구하는 방법  
원의 면적: pi * r ** 2

pi=4*(random한 점이 원내부에 들어올 확률)

10,000개의 난수 순서쌍 (x, y)(x,y) 가운데에는 x^2 + y^2 <=1을 만족하여 원 안의 범위에 포함되는 것들이 있을 것이다.   
그런 순서쌍들의 개수를 세어 전체 난수에 대한 비율을 계산하면 대략적인 원의 면적을 구할 수 있다. 무작위로 뽑힌 난수의 개수가 늘어날수록 더 정확한 결과를 얻을 수 있으나, 그만큼 더 많은 시간이 걸리는 것을 감안


```python
n = 10000
x = np.random.rand(n).round(2)
y = np.random.rand(n).round(2)

x,y
```

```python
plt.figure(figsize = (6,6))
plt.scatter(x, y, s = 3)
```

```python
xx = np.linspace(0,1,100)
plt.plot(xx,(1-xx*xx)**0.5, c='r')  # y^2 = t-x^2
```

```python
z = np.random.rand(500000)
(z<0.5).mean()
```

```python
n = 10000
x = np.random.rand(n).round(2)
y = np.random.rand(n).round(2)
plt.figure(figsize = (6,6))
plt.scatter(x, y, s = 3)

xx = np.linspace(0,1,100)
plt.plot(xx,(1-xx*xx)**0.5, c='r')  # y^2 = t-x^2

pi = ((x**2 + y**2)<1).mean()*4
print(pi)
print(sum((x**2 + y**2) < 1)*4/n)
```

## 람다함수(lambda)  
- 람다함수를 통해 이름이 없는 함수를 만들 수 있다.  
- 람다한수의 장점은 코드의 간결함, 메모리의 절약  
lanbda x : x + 1

```python
fc = lambda x:x+1
fc(10)
```

```python
f = lambda x : x.max()-x.min()

df= pd.DataFrame(np.arange(0,12,1).reshape(4,3))
df
```

```python
f = lambda x : x.max()-x.min()

df= pd.DataFrame(np.arange(0,12,1).reshape(4,3),
                columns=['A','B','C'],index=['a','b','c','d'])                
df
```

```python
df.apply(f,0)  #열기준으로 람다x 최대값-최소값
```

```python
df.apply(f,1)  #행기준으로 람다x 최대값-최소값
```

```python
df.apply(fc)
```
